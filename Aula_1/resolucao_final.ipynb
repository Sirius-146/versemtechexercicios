## Exercício 5

```python
valor_str = float(input("Insira o valor total da conta: "))
valor_str = valor_str.replace(',', '.')
valor = float(valor_str)

gorjeta_str = input("Insira a porcentagem de gorjeta que deseja dar: ")
gorjeta_str = gorjeta_str.replace(',', '.')
gorjeta = float(gorjeta_str)

novo_total = valor + (valor * (gorjeta / 100))
print(f'Com esse valor de gorjeta o total agora é: R${novo_total:.2f}')
```

## Exercício 8

```python
montante_str = input('Digite o montante inicial: ')
montante_str = montante_str.replace(',', '.')
montante = float(montante_str)

taxa_juros_str = input('Digite a taxa de juros ao ano: ')
taxa_juros_str = taxa_juros_str.replace(',', '.')
taxa_juros = float(taxa_juros_str)

anos = int(input('Digite o tempo (em anos): '))

juros_simples = montante * (taxa_juros / 100) * anos

print(f'O valor pago em juros foi: R${juros_simples:.2f}')
```

## Exercício 9

```python
preco_original_str = input('Digite o valor original: ')
preco_original_str = preco_original_str.replace(',', '.')
preco_original = float(preco_original_str)

desconto = float(input('Digite a porcentagem de desconto: '))
novo_preco = preco_original - preco_original * (desconto / 100)

print(f'De: R${preco_original:.2f} \nPor: R${novo_preco:.2f}')
```

## Exercício 10

```python
x1 = int(input('Insira a coordenada x1: '))
y1 = int(input('Insira a coordenada y1: '))
x2 = int(input('Insira a coordenada x2: '))
y2 = int(input('Insira a coordenada y2: '))

distancia = ((x2 - x1)**2 + (y2 - y1)**2)**0.5

print(f'A distância entre os pontos é: {distancia}')
```

## Exercício 11

```python
n1 = int(input('Insira nota 1: '))
p1 = int(input('Insira o peso da nota 1: '))
n2 = int(input('Insira nota 2: '))
p2 = int(input('Insira o peso da nota 2: '))
n3 = int(input('Insira nota 3: '))
p3 = int(input('Insira o peso da nota 3: '))

media_ponderada = ((n1 * p1)+(n2 * p2)+(n3 * p3)) / (p1 + p2 + p3)

print(f'A média ponderada é: {media_ponderada}')
```

## Exercício 12

```python
segundos = int(input('Insira o valor em segundos: '))

QTE_SEG_HORA = 3_600
QTE_SEG_MIN = 60

horas = segundos // QTE_SEG_HORA
resto_horas = segundos % QTE_SEG_HORA

minutos = resto_horas // QTE_SEG_MIN
segundo = resto_horas % QTE_SEG_MIN

print(f'O valor resulta em: {int(horas)} horas, {int(minutos)} minutos e {int(segundo)} segundos.')
```
